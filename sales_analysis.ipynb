# Augment the dataframe with some extra columns to get detail info

## Task 1:
- Create 'Month' column in dataframe for each 'Order Date' 

```python
all_data=all_data[all_data['Order Date'].str[0:2]!='Or']
all_data.head()
```

```python
all_data['Month']=all_data['Order Date'].str[0:2]
all_data['Month']=all_data['Month'].astype('int32')
all_data.head()
```

## Task 2:
- Add 'Sales' column in the dataframe by multiplying 'Price Each' and 'Quantity Ordered'

```python
all_data['Quantity Ordered']=all_data['Quantity Ordered'].astype('int32')
all_data.head()
```

```python
all_data=all_data[all_data['Price Each']!='14.95']
all_data
```

## Convert the required columns to correct data type

```python
all_data['Quantity Ordered']=pd.to_numeric(all_data['Quantity Ordered'])
all_data['Price Each']=pd.to_numeric(all_data['Price Each'])
```

## find Sales for each Order ID
- Ways to do:
    - Multiply 'Quantity Ordered' with 'Price Each'

```python
#all_data.head()
all_data['Sales']=all_data['Quantity Ordered']*all_data['Price Each']
all_data.head()
```

## Reorder the columns in DataFrame

```python
all_data=all_data[['Order ID','Product','Quantity Ordered','Price Each','Sales','Order Date','Month','Purchase Address']]
all_data
```

## Q2. Which city had the highest number of Sales?

```python
all_data.head()
```

### Steps followed:
- First Augment the data by adding new column 'city' using given column 'Purchase Address'
    - Check the data type of 'city' column
    - If there are errors,then clean them
- Secondly, Augment the data by adding new column 'State Code' using given column 'Purchase Address'
- Thirdly Augment the data by adding new column 'Zip Code' using given column 'Purchase Address'
- Fourthly Augment the data by adding new column 'Street' using given column 'Purchase Address'
- First Augment the data by adding new column 'city' using given column 'Purchase Address'
- Then visualize the data with bar chart    

```python
def get_city(address):
    return address.split(',')[1]
all_data['City']=all_data['Purchase Address'].apply(lambda x: get_city(x))
all_data.head()
```

```python
def get_state(address):
    return address.split(', ')[2].split(' ')[0]
all_data['State Code']=all_data['Purchase Address'].apply(lambda x: get_state(x))
all_data.head()
```

```python
def get_zipcode(address):
    return address.split(', ')[2].split(' ')[1]
all_data['Zip Code']=all_data['Purchase Address'].apply(lambda x: get_zipcode(x))
all_data.head()
```

```python
def get_street(address):
    return address.split(', ')[0]
all_data['Street']=all_data['Purchase Address'].apply(lambda x: get_street(x))
all_data.head()
```

```python
all_data
```

```python
all_data.groupby('City').sum()
```

```python
import matplotlib.pyplot as plt
results_2=all_data.groupby('City').sum()
cities=[city for city,df in all_data.groupby('City')]
plt.bar(cities,results_2['Sales'])
plt.xticks(cities,rotation='vertical',size=8)
plt.xlabel('Cities')
plt.ylabel('Sales(in $M)')
plt.show()
# SanFransisco had maximum Sales of more than $8Millions
```

```python
all_data=all_data[['Order ID','Product','Quantity Ordered','Price Each','Sales','Order Date','Month','Hour','Minutes','Seconds',
                   'Purchase Address','City','State Code','Zip Code','Street' 
                  ]]
all_data.head()
```

```python
import matplotlib.pyplot as plt
orders=all_data.groupby(['Hour']).count()
```

```python
hours=[hour for hour,df in all_data.groupby('Hour')]
orders=all_data.groupby(['Hour']).count()
plt.xlabel('Hours')
plt.ylabel('Count of Orders')
plt.plot(hours,orders)
plt.xticks(hours)
plt.grid()
plt.show()

## so we should try to give ads at 11am and 7pm , to increase the likelyhood of Purchase
```
